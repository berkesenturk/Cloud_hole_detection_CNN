#### Labeling For Cloud Hole Dates

```python
import matplotlib.pyplot as plt
import xarray as xr
import pandas as pd

cmap = plt.get_cmap('Greys_r')


labels_path = '../data/processed/julia_labels.csv'
nc_path = '../data/raw/seviri'
```

```python
labels_df = pd.read_csv("../data/raw/labels/labels.csv", index_col=0, )
labels_df = labels_df.sort_index(ascending=True)
labels_df
```

```python
ds = xr.open_dataset(f"{nc_path}/hrv_lr201902.nc")
```

```python
ds.time
```

```python
filtered_data = ds.hrv.sel(time="2019-02-20")

ncols = 4
nrows = (len(filtered_data.time) + ncols - 1) // ncols

fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(80, 80), constrained_layout=True)
axs = axs.ravel()  

for i, time in enumerate(filtered_data.time):
    ax = axs[i]  
    slice_data = filtered_data.sel(time=time)  
    
    slice_data.plot.pcolormesh(
        'lon', 'lat', 
        cmap=cmap,  
        robust=True, 
        ax=ax, 
        add_labels=False,
        cbar_kwargs={
            "label": "Reflectance",
            "pad": 0.04,
            "shrink": 0.75,
            "aspect": 20,
            "extend": "both",
            "format": "%0.1f"
        }
    )
    ax.set_title(str(time.values), fontsize=40) 

for ax in axs[len(filtered_data.time):]:
    ax.set_visible(False)

plt.show()
```
